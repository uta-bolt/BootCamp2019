```python
titanic=pd.read_csv('titanic.csv') 
titanic.info()
```

```python
titanic.pivot_table(values="Survived", columns="Embarked",aggfunc="mean")
```

Those who embarked in C were most likely to survive

```python
titanic.pivot_table(values="Survived",index="Sex", columns="Embarked",aggfunc="mean")
```

Especially women who embarked in C were most likely to survive

```python
titanic.pivot_table(values="Survived",columns="Pclass",aggfunc="mean")
titanic.pivot_table(values="Pclass",columns="Embarked",aggfunc="mean")
```

```python
titanic.pivot_table(values="Survived",columns="Age",aggfunc=["mean","count"])
```

```python
#titanic.assign(agegroup =titanic["Age"])
agegroup =titanic["Age"]
titanic['Agegroup'] = agegroup
titanic.info()
```

```python
titanic.Agegroup[(titanic['Agegroup']< 14)]=1
titanic.Agegroup[(titanic['Agegroup']>=14)&(titanic['Agegroup']<65)]=2
titanic.Agegroup[(titanic['Agegroup']>65)]=3
```

```python
titanic.pivot_table(values="Survived",columns="Agegroup",aggfunc=["mean"])
```

```python
titanic.pivot_table(values="Agegroup",columns="Embarked",aggfunc=["mean"])
```
