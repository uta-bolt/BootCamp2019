# Problem 5

```python
def rosenbrock(x,y):
    res=(1-x)**2+100*(y-x**2)**2
    return res
```

```python
# Create 2 dimensional space:
x = np.linspace(-10, 10, 100)
X, Y = np.meshgrid(x, x)
rb=rosenbrock(X,Y)
```

```python
ax1 = plt.subplot(121)
ax2 = plt.subplot(122)
ax1.pcolormesh(X, Y, rb, cmap="viridis")
ax2.contour(X, Y, rb,[-4000,-1000, 0,1000, 4000], cmap="magma")
ax2.plot(1, 1, 0,'ro', label="minimum")
```
