Problem 5. Compute the following information about the DJIA dataset from Problem 1.
• The single day with the largest gain.
• The single day with the largest loss.
• The month with the largest gain.
• The month with the largest loss.
For the monthly statistics, define the gain (or loss) to be the difference between the DJIA on
the last and first days of the month.

```python
#Calculate the largest gain: 
gain=djia - djia.shift(1)
print("daily gain:",gain.VALUE.idxmax())
print("daily loss:",gain.VALUE.idxmin())
```

```python
djia = djia.set_index(pd.to_datetime(djia.index))
djia_month = djia.resample('M').first()
gainmonth=djia_month - djia_month.shift(1)
print("monthly gain:", gainmonth.VALUE.idxmax())
print("monthly loss:", gainmonth.VALUE.idxmin())
```

```python
ax1 = plt.subplot(121)
djia.plot(color="gray", lw=.3, ax=ax1)
djia.rolling(window=200).mean().plot(color='r', lw=1, ax=ax1)
ax1.legend(["Actual", "Rolling"], loc="lower right")
ax1.set_title("Rolling Average")

ax2 = plt.subplot(122)
djia.plot(color="gray", lw=.3, ax=ax2)
djia.ewm(span=200).mean().plot(color='g', lw=1, ax=ax2)
ax2.legend(["Actual", "EWMA"], loc="lower right")
ax2.set_title("EWMA")
```

```python
ax1 = plt.subplot(121)
djia.plot(color="gray", lw=.3, ax=ax1)
djia.rolling(window=200).min().plot(color='r', lw=1, ax=ax1)
ax1.legend(["Actual", "Rolling"], loc="lower right")
ax1.set_title("Minimum Rolling")

ax1 = plt.subplot(122)
djia.plot(color="gray", lw=.3, ax=ax1)
djia.rolling(window=200).max().plot(color='r', lw=1, ax=ax1)
ax1.legend(["Actual", "Rolling"], loc="lower right")
ax1.set_title("Maximum Rolling")
```
