# Problem 1

```python
import numpy as np
from matplotlib import pyplot as plt
```

```python
def myfun1(n, which):
    n=10
    mu, sigma = 0, 1 # mean and standard deviation
    np.random.seed(1283)
    array=np.random.normal(mu, sigma, (n,n))
    mean=np.mean(array,1) 
    var=np.var(array,1) 
    if which==1:
        return mean 
    else:
        return var
```

```python
myfun1(10,1)
```

```python
def myfun2(x):
    for i in range(100, 1000,100):
        res=np.array(myfun1(i,x))
    return res
```

```python
a=myfun2(1)
b=myfun2(2)
c=np.array(range(100, 1100,100))
```

```python
print(c)
```

```python
plt.plot(c,a,label="mean")
plt.plot(c,b,label="variance")
plt.legend(loc="upper left")
```

# Problem 3

```python
def myfun3(x):
    res=1/(x-1)
    return res
```

```python
b=np.arange(-2,0.999999,0.1)
c=np.arange(1.00001,6,0.1)
```

```python
part1=myfun3(b)
part2=myfun3(c)
```

```python
plt.plot(b,part1,'m--',lw=4)
plt.plot(c,part2,'m--',lw=4)
plt.xlim(-2,6) 
plt.ylim(-6,6)
```

# Problem 4

```python
x=np.arange(0,2*np.pi,0.1)
y1 = np.sin(x)
y2 = np.sin(2*x)
y3 = 2*np.sin(x)
y4 = 2*np.sin(2*x)
```

```python
ax1 = plt.subplot(221)
ax1.plot(x, y1, 'g')
ax1.set_title("sin(x)")
ax2 = plt.subplot(222)
ax2.plot(x, y2,'r--')
ax2.set_title("sin(2x)")
ax3 = plt.subplot(223)
ax3.plot(x, y3,'b--')
ax3.set_title("2sin(x)")
ax4 = plt.subplot(224)
ax4.plot(x, y4, 'm:')
ax4.set_title("2sin(2x)")
plt.axis([0, 2*np.pi, -2, 2])
plt.suptitle("Versions of sin")
```
